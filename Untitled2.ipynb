```python
dfcolumns = ['Week', 'Game_Number',
'points_for', 'points_against', 'times_sacked',
'yards_per_play_offense', 'yards_allowed_per_defensive_play',
'turnovers_lost', 'total_TD', 'FGM', 'penalty_yards1', 'penalty_yards2']

def text_to_num(text):
    return float(text)

def convert_into_float(data):
    data[dfcolumns] = data[dfcolumns].applymap(text_to_num)
    return data
```

```python
dfs2 = convert_into_float(dfs2)
```

```python
dfs2 = dfs2.fillna(value=0.0)
```

```python
dfs2.info()
```

```python
def top_ratio(text):
    return float(text.split(':')[0])/60 + float(text.split(':')[1])/3600
def time_of_possession_to_ratio(data):
    data['time_of_possesion'] = data['time_of_possesion'].apply(lambda x: top_ratio(x))
    return data
```

```python
dfs2 = time_of_possession_to_ratio(dfs2)
```

```python
dfs2.info()
```

```python
dfs2['win1?'] = dfs2['win1?'].apply(lambda x: x[0])
```

```python
dfs2['win1?'] = dfs2['win1?'].replace(to_replace=['W', 'L', 'T'], value=[1.0, 0.0, 0.0])
```

```python
dfs2 = dfs2.sort_values('Week')
```

```python
dfs2['home_away'] = dfs2['home_away'].replace(to_replace=['Home', 'Away'], value=[1,0])
```

```python
dfs2 = dfs2.rename(columns={'home_away':'home'})
```

```python
dfs2 = dfs2.rename(columns={'team1':'team', 'team2':'opp', 'win1?':'win', 
                            'penalty_yards1':'penalty_yards', 
                            'penalty_yards2':'opp_penalty_yards'})
```

```python
dfs2['loss'] = 1 - dfs2['win']
```

```python
dfs2['total_wins'] = dfs2.groupby('team')['win'].cumsum()
```

```python
dfs2['total_losses'] = dfs2.groupby('team')['loss'].cumsum()
```

```python
dfs2['away'] = 1 - dfs2['home']
```
